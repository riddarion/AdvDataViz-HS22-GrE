# Aufgabe 6 a) Einlesen der Daten

```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df = pd.read_csv("data_moods.csv")
df
```

#  Aufgabe 6 b) Datensatz umformatieren

## Umsortieren nach Release Date

```python
#Datensatz nach Release Date aufsteigend sortieren und neu indexieren
dfsort = df.sort_values(by=["release_date"])
dfsort = dfsort.reset_index(drop=True)

dfsort
```

```python
#Um den Index nutzen zu können erstellen wir eine Spalte aus dem Index
dfsort['index1'] = dfsort.index
dfsort
```

## Mood in Zahlenformat umwandeln (später für Parallel Coordinates Farbskala)

```python
#Ursprungsdatensatz für eine weitere Transformation kopieren und Datentyp der Spalte "mood" auf String stellen
dfcolmap = df
dfcolmap= dfcolmap.astype({"mood": str})
dfcolmap
```

```python
#Moods durch Zahlen ersetzen
dfcolmap["mood"].replace({"Happy": 1, "Sad": 2, "Energetic": 3, "Calm": 4}, inplace=True)
dfcolmap
```

## Vorbereiten Jahr für Weiterverwendung

```python
#Ursprungsdatensatz kopieren
dfMoodspYear = df

#Release Datum aufspalten in Jahr, Monat und Tag
dfMoodspYear[["year", "month", "day"]] = dfMoodspYear["release_date"].str.split("-", expand=True)
dfMoodspYear
```

```python
#Von den neuen Spalten behalten wir nur das Jahr
dfMoodspYear = dfMoodspYear.drop(columns=['month', 'day'])
dfMoodspYear
```

```python
#Gruppieren und zählen nach Jahr und Mood
dfMoodspYear = df.groupby(["year", "mood"]).count()
dfMoodspYear
```

```python
#Jahr und Mood wieder nutzbar machen und neuen Index setzen
MpY = dfMoodspYear.iloc[:,0]
MpY = MpY.reset_index(level=(0,1))
MpY["year"] = MpY["year"].astype("int")
MpY = MpY.sort_values(by=["year"])
MpY
```

## Treemap zu Hypothese 5

```python
#Zuerst werden die Songs gezählt
df = pd.read_csv("data_moods.csv")
dfCountMoods = df
dfCountMoods = df.groupby("mood").count()
dfCountMoods["category"] = dfCountMoods.index
dfCountMoods["all"] = "all" # in order to have a single root node
dfCountMoods
```

```python
#Treemap mit manueller Werte-Übergabe, also nicht über den Datensatz
import plotly.graph_objects as go

fig = go.Figure(go.Treemap(
    labels = ["Happy", "Sad", "Energetic", "Calm"],
    values = [140, 197, 154, 195],
    parents = ["Mood", "Mood", "Mood", "Mood"],
    marker_colors = ["red", "yellow", "aqua", "fuchsia"],
    root_color="lightgrey"))

fig.update_layout(width=400, height = 800)
fig.show()
```

# Aufgabe 7 b) Streudiagrammmatrix

```python
#Code zur Streudiagrammmatrix mit allen numerischen Attributen
fig = px.scatter_matrix(df,
                        dimensions=["popularity", "length", "danceability", "acousticness", "energy", "instrumentalness", "liveness", "valence", "loudness", "speechiness", "tempo"],
                        color="mood",
                       title="Streudiagrammmatrix der numerischen Attribute",
                       labels={col:col.replace('_', ' ') for col in df.columns}) #entfert underlines in Titeln
fig.update_traces(diagonal_visible=False, marker_size=1) #Diagramme, die die selben Werte miteinander vergleichen werden ausgeblendet
fig.show()
```

```python
#Streudiagrammmatrix mit Attributen für Hypothesen
#in Aufgabenblatt 3 abgebildet.

df = pd.read_csv("data_moods.csv")
dfMoodspYear = df

dfMoodspYear[["year", "month", "day"]] = dfMoodspYear["release_date"].str.split("-", expand=True)
dfMoodspYear

fig = px.scatter_matrix(dfMoodspYear,
                        dimensions=[ "length", "danceability", "popularity", "year", "mood"],
                        color="mood",
                       title="Streudiagrammmatrix der numerischen Attribute",
                       labels={col:col.replace('_', ' ') for col in df.columns}) #entfert underlines in Titeln
fig.update_traces(diagonal_visible=False, marker_size=2) #Diagramme, die die selben Werte miteinander vergleichen werden ausgeblendet
fig.show()
```
